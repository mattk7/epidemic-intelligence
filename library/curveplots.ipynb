## Functionality
Starting with a BigQuery dataset that looks like [this](https://console.cloud.google.com/bigquery?project=net-data-viz-handbook&ws=!1m5!1m4!4m3!1snet-data-viz-handbook!2ssri_data!3sSIR_0_basins_prevalence)...

### 1. Construct distance matrix

#### Parameters: 

target column (eg. total_infectious, Susceptible_75_plus)

#### Returns:

distance matrix

### 2. Grouping curves
#### Parameters:

distance matrix

grouping method ('abc' or 'mse'), number of clusters (int > 0), number of features (int > 0)
        
#### Returns: 

kmeans results table

### 3. Get MBD
#### Parameters:

original dataset

kmeans results table
        
#### Returns: 

mbd table

<br>

allowing users to directly pull any of those things is going to slow the program down dramatically. it's a graphing library. it should exclusively make graphs. 
<hr>

### 4. Create traditional boxplot
#### Parameters:

Outlier distinction value (eg .9), graphing kwargs

#### Returns:

`Figure` object

### 5. Create functional boxplot

#### Parameters:

centrality metric ('mse', 'abc', or 'mbd'), graphing kwargs

#### Returns:

`Figure` object

### 6. Plot most central curves

#### Parameters:

number of curves, centrality metric, graphing kwargs

#### Returns:

`Figure` object


### Graphing kwargs:

colorlist: list of [CSS named colors](https://www.w3schools.com/cssref/css_colors.php) to be used for graphing different groups

layoutkwargs: keyword arguments to be passed to `plotly`'s [update_layout](https://plotly.com/python/reference/layout/) function
